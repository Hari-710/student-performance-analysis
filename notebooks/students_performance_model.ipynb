```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('/content/StudentsPerformance_cleaned.csv')

df.head()
```

```python
df_ml = df.copy()

labels_cols = ['gender', 'race/ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']
for col in labels_cols:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col])
```

```python
# Defining Features x and y

X = df_ml.drop(['result'], axis=1)
Y = df_ml['result']
```

```python
# Train-Test-Split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = RandomForestClassifier()
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)

print("Accuracy:",accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n",classification_report(Y_test, y_pred))
```
